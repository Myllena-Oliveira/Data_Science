```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tecent',
            "McDonald's", 'Visa',
            'Facebook', 'Aibaba',
            'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           12.273]
```

```python
plt.plot (empresas, bilhoes)
```

```python
plt.plot (empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot (empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel ('Empresas')
plt.ylabel ('Bilhes')
```

Usando o plt.show para evitar a sada de texto

```python
plt.plot (empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel ('Empresas')
plt.ylabel ('Bilhes')
plt.show()
```

Redimensionando

```python
plt.rcParams["figure.figsize"] = (15,8)
plt.plot (empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel ('Empresas')
plt.ylabel ('Bilhes')
plt.show()
```

Tamanho da fonte

```python
plt.plot (empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
plt.show()
```

Mudando a cor e legendando

```python
plt.plot (empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

Aumentando a espessura da linha

```python
plt.plot (empresas, bilhoes, color = 'green', lw = 10)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

Adicionando marcador

```python
plt.plot (empresas, bilhoes, color = 'green', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

Adicionando as linhas de grade

```python
plt.plot (empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.grid(True)
plt.show()
```

Mudando o tipo do grfico e colocando a legenda em cima

```python
plt.bar (empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

Gerando grfico mais simples

```python
plt.bar (empresas,bilhoes)
```

Colorindo os grficos

```python
cores = ['black', 'dimgray', 'dimgrey', 'gray', 'grey',
         'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

Lista de opes para estilizar os grficos

```python
plt.style.available
```

```python
plt.style.use('ggplot')
plt.bar(empresas, bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
#plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

```python
plt.style.use('dark_background')
plt.bar(empresas, bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
#plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

```python
plt.style.use('seaborn-dark')
plt.barh(empresas, bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
#plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```

```python
plt.style.use('seaborn-dark')
plt.scatter(empresas, bilhoes, color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel ('Empresas', fontsize = 20)
plt.ylabel ('Bilhes', fontsize = 20)
#plt.legend(['Bilhes de dlares'], loc = 'right', fontsize = 15)
plt.show()
```
